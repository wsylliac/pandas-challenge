## PyCity Schools Analysis 



```python
# Dependecies and Setup
import pandas as pd
import numpy as np

# File to load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

#Read School and student Data file and store into Pandas Data Frame 
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#Combine the data into a single dataset
school_data_to_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_to_complete.head()
```

## District Summary

Perform the necessary calculations and then create a high-level snapshot of the district's key metrics in a DataFrame.

Include the following:

* Total number of unique schools
* Total students
* Total budget
* Average math score
* Average reading score
* % passing math (the percentage of students who passed math)
* % passing reading (the percentage of students who passed reading)
* % overall passing (the percentage of students who passed math AND reading)

```python
# Calculating the total number of schools
total_schools=school_data["school_name"].count()
total_schools

#Calculating the number of students
total_students=student_data["student_name"].count()
total_students

#Calculating the total budget
total_budget=school_data["budget"].sum()
total_budget

#Calculating the average of the math scores
avg_math=student_data[("math_score")].mean()
avg_math

#Calculating the average of the reading scores
avg_reading=student_data[("reading_score")].mean()
avg_reading

#Calculating the overall passing rate such as the math score + reading score/2
overall_avg=((avg_math + avg_reading) /2)

#Calculating the percentage of students that passed with a math score of 70 or higher 
perc_math_pass=(school_data_to_complete["math_score"]>70).sum() / school_data_to_complete["math_score"].count()*100
perc_math_pass

#* Calculating the percentage of students that passed with a reading score of 70 or higher
perc_reading_pass=(school_data_to_complete["reading_score"]>70).sum() / school_data_to_complete["reading_score"].count()*100
perc_reading_pass

##(student_data["reading_score"]>70).count() *100
```

```python
# Summary of the District Summary
District_Summary = pd.DataFrame({"Total Schools":[total_schools], "Total Students":[total_students], "Total Budget" 
                                 "Average Math Score":[avg_math], "Average Reading Score": [avg_reading], "% Passing"
                                 "% Passing Reading":[perc_math_pass],"%o Overall Passing Rate":[overall_avg]})
District_Summary
```

```python
# Creating a diff DF for District Summary to do mapping
District_Summary_DF=pd.DataFrame(District_Summary, columns=["Total Schools","Total Students","Total Budget",
                                                            "Average Math Score","Average Reading Score","%Passing"
                                                            "%Overall Passing Rate"])
District_Summary_DF
```

```python
#mapping the District_summary_Df

District_Summary_DF["Total Students"]=District_Summary_DF["Total Students"].map("{:,}".format)
District_Summary_DF["Total Budget"]=District_Summary_DF["Total Budget"].map("{:,}".format)

District_Summary_DF["Average Math Score"]=District_Summary_DF["Average Math Score"].map("{:,.2f}".format)
District_Summary_DF
```

### Top Performing Schools (By Passing Rate)

```python
sch_count= len(school_data_complete["school_name"].unique())
sch_count
###school_count= school_data_complete["school_name"].unique()
```

```python
#School Name, setting up the school name and the type as index
sch_types=school_data.set_index(["school_name"])["type"]
#sch_types.head()
```

```python
#The total of students in each school
stu_per_sch=school_data_to_complete["school_name"].value_counts()
stu_per_sch.head()
```

```python
#The total school budget
sch_budget=school_data_to_complete.groupby(["school_name"])["budget"].mean()
sch_budget.head()
```

```python
#Per student budget
per_stu_bud=sch_budget/stu_per_sch
per_stu_bud
```

```python
#The Average Math Score for Each School
avg_math_perSch=school_data_to_complete.groupby(["school_name"])["math_score"].mean()
avg_math_perSch.head()
```

```python
#The Average Reading Score for Each School
avg_red_perSch= school_data_to_complete.groupby(["school_name"])["reading_score"].mean()
avg_red_perSch.head()
```

```python
# The Percent of Passing Math per School
pas_math_scor=school_data_to_complete.loc[school_data_to_complete["math_score"]>=70]
pas_math_scor

#Assigning the math score per school
group_math_sch=pas_math_scor["school_name"].value_counts()
group_math_sch.head()

#Final Percentage Results
percent_math=group_math_sch/stu_per_sch*100
percent_math
```

```python
# The Percent Passing Reading per School
read_scor=school_data_to_complete.loc[school_data_to_complete["reading_score"]>=70]

#Assigning the reading score per school
by_read_PerSch= read_scor["school_name"].value_counts()
by_read_PerSch.head()

#Final Percentage Results
percent_read=by_read_PerSch/stu_per_sch*100
percent_read
```

```python
# The percent of the overall passing rate
overall=percent_math + percent_read/stu_per_sch
overall
```

```python
#Create a new DF for the School Summary
school_summary_df = pd.DataFrame({"School Type":sch_types, "Total Students":stu_per_sch, "Total School Budget":sch_budget
                                ,"Per Student Budget":per_stu_bud ,"Average Math Score" :avg_math_perSch,
                             "Average Reading Score":avg_red_perSch,
                             "% Passing Math":percent_math, "% Passing Reading":percent_read,
                            "% Overall Passing Rate":overall})
school_summary_df.head()
```

```python
#The top performing schools according to the passing rate
TopSch_df=school_summary_df.sort_values(["% Passing Math","% Passing Reading","% Overall Passing Rate" ], 
                                        ascending=False)
TopSch_df.head()
```

```python
#Mapping the top performing schools by the passing rate
TopSch_df["Total School Budget"]=TopSch_df["Total School Budget"].map("${:,.2f}".format)
TopSch_df["Per Student Budget"]=TopSch_df["Per Student Budget"].map("${:,.2f}".format)

TopSch_df
```
